```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('DSL-StrongPasswordData.csv')
cols = df.columns

dd_cols = [col for col in cols if col.startswith('DD')]
```

```python
dfs_subj = [df[df.subject == subj] for subj in np.unique(df.subject)]
```

```python
key_press_times = []
for df_subj in dfs_subj:
    key_press_times_ = np.array(df_subj[dd_cols])
#     for csession in range(key_press_times_.shape[0]):
#         key_press_times_[csession, :] /= np.sum(key_press_times_[csession, :])
    key_press_times.append(key_press_times_)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(15, 10), tight_layout=True)
for ax, df_subj, key_press_times_ in zip(axs, dfs_subj, key_press_times):
    for csession in range(len(key_press_times_)):
        ax.scatter(key_press_times_[csession].cumsum(), csession*np.ones(len(key_press_times_[csession])), c='k', marker='|')
        
    for col_1 in range(key_press_times_.shape[1]-1):
        col_2 = col_1 + 1
        print(stats.pearsonr(key_press_times_[:, col_1], key_press_times_[:, col_2])[0])
```
