```python
import matplotlib.pyplot as plt
import numpy as np

from disp import set_plot

N = 1000
DT = .001
TAU = .02

J = 10*np.diag(np.ones(N-1), -1)
```

```python
t = np.arange(0, 3, DT)

BETA = 10
TH = 1
phi = lambda h: (1 + np.tanh(BETA*(h-TH)))/2
phi = lambda h: h  #(1 + np.tanh(BETA*(h-TH)))/2

hs = np.nan*np.zeros((len(t), N))
hs[0, :] = 10*np.eye(N)[0, :]

rs = np.nan*np.zeros((len(t), N))
rs[0, :] = phi(hs[0, :])

for ct, t_ in enumerate(t[1:], 1):
    dh = (DT/TAU)*(-hs[ct-1, :] + J@rs[ct-1, :])
    hs[ct, :] = hs[ct-1, :] + dh
    rs[ct, :] = phi(hs[ct, :])
```

```python
PLOT_EVERY = 10
fig, axs = plt.subplots(2, 1, figsize=(15, 5), tight_layout=True)
axs[0].plot(t, rs[:, 10:20])
axs[1].plot(t, hs[:, 10:20])

set_plot(axs[0], y_lim=(-.1, 1.1), x_label='Time (s)', y_label='r')
set_plot(axs[1], x_label='Time (s)', y_label='h')
```
